```
%matplotlib inline
import math
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
DATA_ROOT = '/content/drive/My Drive/Final-Project-for-CS109a/data_prelim/'
```

**Loading of Boston DataSets**

```
boston_crime_covid_df = pd.read_csv(DATA_ROOT + 'boston_crime_covid_yolanda.csv').set_index(['INCIDENT_NUMBER','OFFENSE_CODE']).drop(columns=["Unnamed: 0","SHOOTING"])
boston_crime_covid_df.head()
```

```
boston_crime_socio_eco_df = pd.read_csv(DATA_ROOT + 'boston_crime_james.csv').set_index(['INCIDENT_NUMBER','OFFENSE_CODE'])
boston_crime_socio_eco_df.head()
```

```
boston_elws_df = pd.read_csv(DATA_ROOT + 'boston_election_locations_weather_sunset_ramzi_final.csv').set_index(['INCIDENT_NUMBER','OFFENSE_CODE']).drop(columns=["Unnamed: 0","Unnamed: 0.1"])
boston_elws_df.head()
```

```
print(boston_crime_covid_df.shape)
print(boston_crime_socio_eco_df.shape)
print(boston_elws_df.shape)
```

**Loading New York City Data Sets**

```
new_york_crime_covid_df = pd.read_csv(DATA_ROOT + 'new_york_crime_covid_yolanda.csv').set_index('CMPLNT_NUM').drop(columns=["Unnamed: 0"])
new_york_crime_covid_df.head()
```

```
new_york_crime_socio_eco_df = pd.read_csv(DATA_ROOT + 'nyc_crime_local_prop_values.csv').set_index('CMPLNT_NUM')
new_york_crime_socio_eco_df.head()
```

```
new_york_elws_df = pd.read_csv(DATA_ROOT + 'nyc_election_locations_weather_sunset_ramzi_final.csv').set_index(["CMPLNT_NUM"]).drop(columns=["Unnamed: 0"])
new_york_elws_df.head()
```

```
print(new_york_crime_covid_df.shape)
print(new_york_crime_socio_eco_df.shape)
print(new_york_elws_df.shape)
```

**Boston Data Merge and Cleansing**

```
crime_covid_df = boston_crime_covid_df[["DISTRICT","YEAR","MONTH","DAY_OF_WEEK","HOUR","LATITUDE","LONGITUDE","COVID_PERIOD","VIOLENT_CRIME","OFFENSE_TYPE"]]
crime_socio_eco_df = boston_crime_socio_eco_df[["WHITE","BLACK","HISPANIC","ASIAN","OTHER","16-19-EMPLOYMENT","20-34-EMPLOYMENT","MEDIAN_INCOME","GEOHASH","LOCAL_PROP_VALUES"]]
boston_elws_df.dropna(subset=['DISTRICT'], inplace=True)
crime_elws_df = boston_elws_df[["Streetlight_density","Police_Station_Distance","Hospitals_Distance","College_Univ_Distance","election","PRCP","SNOW","NIGHT"]]
boston_crime_analysis_data_df = crime_covid_df.join(crime_socio_eco_df).join(crime_elws_df)
boston_crime_analysis_data_df.head()
```

```
#dropping NA / -1 lat lon values
boston_crime_analysis_data_df.dropna(subset=['LATITUDE','LONGITUDE'], inplace=True)
boston_crime_analysis_data_df = boston_crime_analysis_data_df[(boston_crime_analysis_data_df["LATITUDE"] != -1) & (boston_crime_analysis_data_df["LONGITUDE"] != -1) ]

# dropping External district due to missing socio-economic data
boston_crime_analysis_data_df = boston_crime_analysis_data_df[boston_crime_analysis_data_df["DISTRICT"] != "External"]
```

**New York City Data Merge and Cleansing**

```
crime_covid_df = new_york_crime_covid_df[["BOROUGH","YEAR","MONTH","DAY_OF_WEEK","HOUR","Latitude","Longitude","COVID_PERIOD","VIOLENT_CRIME","OFFENSE_TYPE","GEOHASH"]]
crime_socio_eco_df = new_york_crime_socio_eco_df[["LOCAL_PROP_VALUES"]]
crime_elws_df = new_york_elws_df[["Streetlight_density","Police_Station_Distance","Hospitals_Distance","College_Univ_Distance","election","PRCP","SNOW","NIGHT"]]
new_york_crime_analysis_data_df = crime_covid_df.join(crime_socio_eco_df).join(crime_elws_df)
new_york_crime_analysis_data_df.head()
```

```
#dropping NA / -1 lat lon values
new_york_crime_analysis_data_df.dropna(subset=['Latitude','Longitude'], inplace=True)
new_york_crime_analysis_data_df = new_york_crime_analysis_data_df[(new_york_crime_analysis_data_df["Latitude"] != -1) & (new_york_crime_analysis_data_df["Longitude"] != -1) ]
```
